```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

#step1.....generate_sample__data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)


#step2.....Try_different_cluster_counts
scores = []
for k in range(2, 8):  # K must be >= 2
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)
    print(f"K={k}, Silhouette Score={score:.3f}")

#step3....Plot_scores
plt.plot(range(2, 8), scores, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for KMeans Clustering")
plt.show()

```

Kmeans Clustering

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# #step1...generate/_sample_ddata
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# #steep2.._set_clusters_to_predict
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)  # Predict cluster labels for each point

# step3....Calculate Silhouette Score
silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score for K-Means:", silhouette_avg)

# ---------- Visualize Clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200, label='Centroids')

plt.title(f"K-Means Clustering (Silhouette Score: {silhouette_avg:.2f})")
plt.legend()
plt.show()
```

Hierarchiel CLustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Create dataset
X, _ = make_blobs(n_samples=100, centers=3, cluster_std=1.0, random_state=42)

# Generate linkage matrix and plot dendrogram
linked = linkage(X, method='ward')
plt.figure(figsize=(8, 5))
dendrogram(linked)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = agg.fit_predict(X)

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.title("Agglomerative Hierarchical Clustering")
plt.show()
```
